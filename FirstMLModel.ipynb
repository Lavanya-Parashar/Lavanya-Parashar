# **MY FIRST ML PROJECT**

## **Load Data**


```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv')
df
```

### Data Preparation



```python
y=df['logS']
y
```

```python
x=df.drop('logS',axis=1)
x
```

### Data Splitting

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=100)
```

```python
x_train
```

```python
x_test
```

## **Linear Rgression**

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train, y_train)
```

##**Applying the model to make predictions**





```python
y_lr_train_pred=lr.predict(x_train)
y_lr_test_pred=lr.predict(x_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

## **Evaluation of Model Performace**

```python
y_train
```

```python
y_lr_train_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse=mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2=r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print('LR MSE (Train):',lr_train_mse)
print('LR R2 (Train):',lr_train_r2)
print('LR MSE (Test):',lr_test_mse)
print('LR R2 (Test):',lr_test_r2)
```

```python
lr_train_r2
```

```python
lr_test_r2
```

```python
lr_results = pd.DataFrame([['Linear Regression',lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]],
                          columns=['Method','Training MSE','Training R2','Test MSE','Test R2'])
```

```python
lr_results
```

## **Random Forest**

## **Training the model**

```python
#Training the model
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(x_train, y_train)
```

### **Applying the model to make predictions**

```python
#Applying the model to make predictions
y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
```

### **Evaluate model performance**

```python
#Evaluate model performance
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_results=pd.DataFrame(['RandomForest', lr_train_mse, lr_test_r2, lr_test_mse, lr_test_r2]).transpose()
rf_results.columns = ['Method', 'Training MSE','Training R2','Test MSE','Test R2']
rf_results
```

## **MODEL COMPARISON**

```python
df_models = pd.concat([lr_results, rf_results], axis=0).reset_index()
df_models.drop(columns=['index'], inplace=True)
df_models
```

## **DATA VISUALIZATION**

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(5,5))
plt.scatter(x=y_train, y=y_lr_train_pred, c='#e63e62' , alpha=0.3)

z=np.polyfit(y_train, y_lr_train_pred, 1)
p=np.poly1d(z)

plt.plot(y_train, p(y_train), '#3c1414')
plt.xlabel('Experimental logS')
plt.ylabel('Predicted logS')
plt.show()
```
